# 1464. Maximum Product of Two Elements in an Array

```python
nums = [3, 4, 5, 2]

# find two largest numbers
nums.sort()

max_product = (nums[-1] - 1) * (nums[-2] - 1)

print(max_product)
```

```python
class Solution:
    def maxProduct(self, nums):
        nums.sort()
        return (nums[-1] - 1) * (nums[-2] - 1)
```

# 1395. Count Number of Teams

```python
rating = [2, 5, 3, 4, 1]
n = len(rating)
count = 0

for j in range(n):
    left_smaller = left_greater = 0
    right_smaller = right_greater = 0

    for i in range(j):
        if rating[i] < rating[j]:
            left_smaller += 1
        else:
            left_greater += 1

    for k in range(j + 1, n):
        if rating[k] < rating[j]:
            right_smaller += 1
        else:
            right_greater += 1

    count += left_smaller * right_greater
    count += left_greater * right_smaller

print(count)
```

```python
class Solution:
    def numTeams(self, rating):
        n = len(rating)
        count = 0

        for j in range(n):
            ls = lg = rs = rg = 0

            for i in range(j):
                if rating[i] < rating[j]:
                    ls += 1
                else:
                    lg += 1

            for k in range(j + 1, n):
                if rating[k] < rating[j]:
                    rs += 1
                else:
                    rg += 1

            count += ls * rg + lg * rs

        return count
```

# 1450. Number of Students Doing Homework at a Given Time

```python
startTime = [1,2,3]
endTime = [3,2,7]
queryTime = 4

count = 0
for i in range(len(startTime)):
    if startTime[i] <= queryTime <= endTime[i]:
        count += 1

print(count)
```

```python
class Solution:
    def busyStudent(self, startTime, endTime, queryTime):
        count = 0
        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1
        return count
```

# 1342. Number of Steps to Reduce a Number to Zero

```python
num = 14
steps = 0

while num > 0:
    if num % 2 == 0:
        num //= 2
    else:
        num -= 1
    steps += 1

print(steps)
```

```python
class Solution:
    def numberOfSteps(self, num):
        steps = 0
        while num > 0:
            if num % 2 == 0:
                num //= 2
            else:
                num -= 1
            steps += 1
        return steps
```

# 338. Counting Bits

```python
n = 5
ans = []

for i in range(n + 1):
    ans.append(bin(i).count("1"))

print(ans)
```

```python
class Solution:
    def countBits(self, n):
        ans = [0] * (n + 1)
        for i in range(1, n + 1):
            ans[i] = ans[i >> 1] + (i & 1)
        return ans
```
