# 1295. Find Numbers with Even Number of Digits

```python
address = "1.1.1.1"

def defangIPaddr(address):
    return address.replace(".", "[.]")

print(defangIPaddr(address))
```

# 1512. Number of Good Pairs

```python
nums = [1, 2, 3, 1, 1, 3]

freq = {}
count = 0

for num in nums:
    if num in freq:
        count += freq[num]
        freq[num] += 1
    else:
        freq[num] = 1

print(count)
```

```python
class Solution:
    def numIdenticalPairs(self, nums):
        freq = {}
        count = 0

        for num in nums:
            if num in freq:
                count += freq[num]
                freq[num] += 1
            else:
                freq[num] = 1

        return count
```

# 1365. How Many Numbers Are Smaller Than the Current Number

```python
nums = [8, 1, 2, 2, 3]
result = []

for i in range(len(nums)):
    count = 0
    for j in range(len(nums)):
        if nums[j] < nums[i]:
            count += 1
    result.append(count)

print(result)
```

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums):
        result = []

        for i in range(len(nums)):
            count = 0
            for j in range(len(nums)):
                if nums[j] < nums[i]:
                    count += 1
            result.append(count)

        return result
```

# 1281. Subtract the Product and Sum of Digits of an Integer

```python
n = 234

product = 1
sum_digits = 0

while n > 0:
    digit = n % 10
    product = product * digit
    sum_digits = sum_digits + digit
    n = n // 10

print(product - sum_digits)
```

```python
class Solution:
    def subtractProductAndSum(self, n):
        product = 1
        sum_digits = 0

        while n > 0:
            digit = n % 10
            product *= digit
            sum_digits += digit
            n //= 10

        return product - sum_digits
```

# 1486. XOR Operation in an Array

```python
n = 5
start = 0

xor_result = 0

for i in range(n):
    xor_result ^= (start + 2 * i)

print(xor_result)
```

```python
class Solution:
    def xorOperation(self, n, start):
        xor_result = 0

        for i in range(n):
            xor_result ^= (start + 2 * i)

        return xor_result
```
